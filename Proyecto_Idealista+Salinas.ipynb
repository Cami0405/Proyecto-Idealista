#### ¿Las amenidades como piscina, jardín y aire acondicionado están asociadas con precios más altos y un mayor número de baños?

Hipótesis: Las propiedades con piscina, jardín y aire acondicionado tienen precios más altos.
Variables a usar:
PRICE (cuantitativa, eje Y): Precio de la propiedad.
HASSWIMMINGPOOL, HASGARDEN, HASAIRCON (cualitativas/binarias, para categorizar): Indicadores de presencia de piscina, jardín y aire acondicionado (1 = sí, 0 = no).

Primero vamos a crear una nueva variable categórica que indique si una propiedad tiene todas, algunas o ninguna de estas amenidades.
Crearemos una columna que clasifique las propiedades según si tienen piscina, jardín y aire acondicionado:

Todas: HASSWIMMINGPOOL=1, HASGARDEN=1, HASAIRCON=1.
Algunas: Al menos una de las tres, pero no todas.
Ninguna: HASPOOL=0, HASGARDEN=0, HASAIRCON=0.

```python
# Creamos una nueva columna para clasificar las propiedades según las amenidades
def classify_amenities(row):
    if row['HASSWIMMINGPOOL'] == 1 and row['HASGARDEN'] == 1 and row['HASAIRCONDITIONING'] == 1:
        return 'Todas'
    elif row['HASSWIMMINGPOOL'] == 0 and row['HASGARDEN'] == 0 and row['HASAIRCONDITIONING'] == 0:
        return 'Ninguna'
    else:
        return 'Algunas'

df_idealista['AMENITIES'] = df_idealista.apply(classify_amenities, axis=1)
```

```python
# Creamos un boxplot con Seaborn
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_idealista, x='AMENITIES', y='PRICE', order=['Ninguna', 'Algunas', 'Todas'])

# Añadimos títulos y etiquetas
plt.title("Distribución de Precios según la Presencia de Amenidades (Piscina, Jardín, Aire Acondicionado)")
plt.xlabel("Presencia de Amenidades")
plt.ylabel("Precio (€)")

# Guardamos el gráfico como imagen
plt.savefig("price_vs_amenities_boxplot.png")
```

Se observa que la mediana de "Todas" es mayor que la de "Algunas" o "Ninguna" amenidad, lo cual significa que evidentemente encarecen el precio de las viviendas. 

Se trata de eliminar outliers de la variable PRICE usando la técnica de los cuartiles (también conocida como el método del rango intercuartílico o IQR)

1. Calcular los cuartiles:
Q1 (primer cuartil, percentil 25): 25% de los datos están por debajo de este valor.
Q3 (tercer cuartil, percentil 75): 75% de los datos están por debajo de este valor.
IQR (rango intercuartílico): Q3 - Q1.
2. Definir los límites:
Límite inferior: Q1 - 1.5 * IQR.
Límite superior: Q3 + 1.5 * IQR.
Los valores fuera de estos límites se consideran outliers.
3. Filtrar los datos:
Eliminar las filas donde PRICE esté fuera de los límites.

#### Cálculo de IQR:
##### quantile(0.25) y quantile(0.75) calculan Q1 y Q3.
IQR = Q3 - Q1 calcula el rango intercuartílico.
* Límites:
lower_bound = Q1 - 1.5 * IQR y upper_bound = Q3 + 1.5 * IQR definen el rango aceptable.
* Filtrado:
df_no_outliers = df[(df_idealista['PRICE'] >= lower_bound) & (df['PRICE'] <= upper_bound)] # se crea un nuevo DataFrame sin los outliers.
* Boxplot:
Usamos el DataFrame filtrado (df_no_outliers) para generar el boxplot.

```python
# Paso 1: Calcular Q1, Q3 e IQR para PRICE
Q1 = df_idealista['PRICE'].quantile(0.25)
Q3 = df_idealista['PRICE'].quantile(0.75)
IQR = Q3 - Q1

# Paso 2: Definir los límites para identificar outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Paso 3: Filtrar el DataFrame para eliminar outliers en PRICE
df_no_outliers = df_idealista[(df_idealista['PRICE'] >= lower_bound) & (df_idealista['PRICE'] <= upper_bound)]
df_no_outliers['AMENITIES'] = df_no_outliers.apply(classify_amenities, axis=1)

# Crear un boxplot con Seaborn usando los datos sin outliers
plt.figure(figsize=(8,4))
sns.boxplot(data=df_no_outliers, x='AMENITIES', y='PRICE', order=['Ninguna', 'Algunas', 'Todas'])

# Añadir títulos y etiquetas
plt.title("Distribución de Precios según Amenidades (Sin Outliers)")
plt.xlabel("Presencia de Amenidades")
plt.ylabel("Precio (€)")

# Guardar el gráfico como imagen
plt.savefig("price_vs_amenities_boxplot_no_outliers.png")
```

El nuevo boxplot no muestra puntos extremos (outliers) en la variable PRICE, lo que hace que las cajas y los bigotes reflejen mejor la distribución central de los datos.

Se mantiene la demostración de la Hipótesis, es decir, la mediana de precios en la categoría "Todas" (propiedades con piscina, jardín y aire acondicionado) sigue siendo más alta que en "Algunas" y "Ninguna", por lo tanto, las amenidades aumentan el precio. 

La eliminación de outliers evita que valores extremos distorsionen las medianas o los rangos.
Comparado con el gráfico anterior (con outliers), las cajas podrían ser más compactas, y las diferencias entre categorías podrían ser más claras.
